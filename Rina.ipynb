# Imports

```python
from collections import Counter
import copy
import json

import numpy as np
import pandas as pd

from IPython.display import display, HTML
```

# Function Definitions

```python
def build_basic_df(name, data):
    data = copy.deepcopy(data)
    keys = ["p", "r", "f", "counts"]
    data = {
        key:value
        for key,value in data.items()
        if key in keys
    }
    data.update(data["counts"])
    data.pop("counts")
    return pd.DataFrame(data.values(), index=data.keys(), columns=[name])

def convert_to_df(data_dict):
    df_list = list()
    for key,value in data_dict.items():
        tmp_df_list = list()
        name = f"OVERALL"
        data = copy.deepcopy(value)
        data.pop("score_per_type")
        tmp_df_list.append( build_basic_df(name, data) )
        
#         print( json.dumps(value, indent=4) )
        if "score_per_type" in value and "ent_type" in value["score_per_type"]:
            for tmp_key,tmp_value in value["score_per_type"]["ent_type"].items():
                name = f"{tmp_key}"
                data = tmp_value
                tmp_df_list.append( build_basic_df(name, data) )
        df = pd.concat(tmp_df_list, axis=1)
        df.columns = pd.MultiIndex.from_product([[key], df.columns])
        df_list.append(df)
    return pd.concat(df_list, axis=1).astype(float)
```
