# Data Loading

```python
df1 = pd.read_excel("CMOR.xlsx", engine='openpyxl',)
df2 = pd.read_excel("COE.xlsx", engine='openpyxl',)

df1.info(verbose=True)
display(df1)

df2.info(verbose=True)
display(df2)
```

```python
def compare_table(
    df1,
    df2,
    col_list=None,
    groupby=["Transaction_mth","Segment"],
    title=None,
    save=None
):
    assert title is None or len(title) == 2, "expecting exactly 2 titles"
    if col_list is None:
        col_list = set.intersection(*[
            set(df.columns) - set(groupby)
            for df in [df1, df2]
        ])
    if title is None:
        title = ["Table 1", "Table 2"]
    merge_df = pd.concat(
        [
            df.set_index(groupby)[col_list]
            for df in [df1, df2]
        ],
        keys=title,
        axis=1
    )
    for column in col_list:
        merge_df[("Variation", column)] = (df1[column] - df2[column]).values / 100.
    if save is not None:
        merge_df.to_excel(save)
    return merge_df
    
display(
    compare_table(
        df1,
        df2,
        title=["COE", "CMoR"],
        save="test.xlsx"
    )
)
```
