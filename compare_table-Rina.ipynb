# Data Loading

```python
df1 = pd.read_excel("data1.xlsx", engine="openpyxl")[:-4].copy()
df2 = pd.read_excel("data2.xlsx", engine="openpyxl").dropna()

df1.info(verbose=True)
display(df1)

df2.info(verbose=True)
display(df2)
```

```python
def compare_table(
    df1,
    df2,
    col_list=None,
    groupby=["Transaction_mth","Segment"],
    title=["Table 1", "Table 2"],
    save=None
):
    assert title is None or len(title) == 2, "expecting exactly 2 titles"
    if col_list is None:
        col_list = set.intersection(set(df1.columns), set(df2.columns)) - set(groupby)
    
    df1 = df1.set_index(groupby)[col_list]
    df2 = df2.set_index(groupby)[col_list]
    index = [
        item
        for item in df1.index
        if item in df2.index
    ]
    variation_df = (df1.loc[index,col_list] - df2.loc[index,col_list]) / 100.
    
    merge_df = pd.concat(
        [
            df.loc[index,col_list]
            for df in [df1, df2, variation_df]
        ],
        keys=title+["Variation"],
        axis=1
    )
    display(merge_df)
    if save is not None:
        merge_df.to_excel(save)
    return merge_df
    
display(
    compare_table(
        df1,
        df2,
        title=["COE", "CMoR"],
        save="test.xlsx"
    )
)
```
